## Group by and Aggregations

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').agg({'Flavor': ['max','count','sum'], 'Texture Rating':['max','count','sum']})
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor': ['max','count','sum']})
```

```python
df.groupby('Base Flavor').describe()
```

## Merge, Join and Concatenate

```python
df1= pd.read_csv(r"C:\Users\ViztaEdu\Desktop\Data Analyst Program\python\LOTR.csv")
df1
```

```python
df2= pd.read_csv(r"C:\Users\ViztaEdu\Desktop\Data Analyst Program\python\LOTR 2.csv")
df2
```

```python
#merge

#inner join
df1.merge(df2,how = 'inner', on =['FellowshipID','FirstName'])

```

```python
#outer join

df1.merge(df2,how = 'outer')
```

```python
#left join

df1.merge(df2,how = 'left')
```

```python
#right join

df1.merge(df2,how = 'right')
```

```python
#Cross join

df1.merge(df2,how = 'cross')
```

```python
df1.join(df2,on = 'FellowshipID',how ='outer', lsuffix='_Left',rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left',rsuffix='_Right', how ='outer')
df4
```

```python
# COncatenate

pd.concat([df1,df2],join ='outer')
```
