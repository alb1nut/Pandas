## FIltering and Ordering In Pandas


```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\ViztaEdu\Desktop\Data Analyst Program\python\world_population.csv")
df
```

```python
df[df['Rank'] < 10]
```

```python
specific_countries = ['Bangladesh','Brazil']
```

```python
#to check if an array of data is in a dataframe

df[df['Country'].isin(specific_countries)]
```

```python
#to search for a specific string value

df[df['Country'].str.contains('United')]
```

```python
#Setting index

df2 = df.set_index('Country')
df2
```

```python
#filtering items
#filtering vertically
df2.filter(items= ['Continent','CCA3'])
```

```python
#filtering horizontally

df2.filter(items= ['Zimbabwe'],axis=0)
```

```python
#searching through with exact index name

df2.loc['United States']
```

```python
#searching through with the index of the location
df2.iloc[3]
```

```python
df
```

```python
#sorting data 

df[df['Rank'] < 10].sort_values(by='Rank')
```

```python
#multiple coulmns sort

df[df['Rank'] < 10].sort_values(by=['Rank', 'Country'], ascending =False)
```
